**Data Pre-processing**

Data preprocessing is an essential step in data analysis. It involves cleaning and transforming raw data into a structured format suitable for analysis.

In this stage, missing values are handled, duplicate records are removed, data types are corrected, and new features are created to improve the quality of the dataset.

**Handling Missing Values**

```python
df.isnull().sum()
```



*   This function is used to identify missing values in each column of the dataset.


*   From the output, it was observed that several columns contain missing values such as age, years_of_experience, hourly_rate (USD), rating, is_active, and client_satisfaction.



**Filling numerical columns**

```python
df["age"].fillna(df["age"].median(), inplace=True)
```

```python
df["years_of_experience"].fillna(df["years_of_experience"].median(), inplace=True)
```

```python
df["rating"].fillna(df["rating"].median(), inplace=True)
```



*   Missing values in numerical columns such as **age, years_of_experience, and rating** were replaced using the median value.

*   The median is preferred over the mean because it is less affected by extreme values (outliers), making it a reliable measure for data imputation.



**Cleaning and Standardizing Columns**

Cleaning and standardizing columns ensures that the dataset is consistent, accurate, and ready for analysis by removing errors, handling missing values, and making the data easier to interpret.

**Hourly Rate (USD)**

```python
df["hourly_rate (USD)"] = df["hourly_rate (USD)"].astype(str)
df["hourly_rate (USD)"] = df["hourly_rate (USD)"].str.replace("USD","")
df["hourly_rate (USD)"] = df["hourly_rate (USD)"].str.replace("$","")
df["hourly_rate (USD)"] = pd.to_numeric(df["hourly_rate (USD)"], errors="coerce")
df["hourly_rate (USD)"].fillna(df["hourly_rate (USD)"].median(), inplace=True)
```




*   The hourly_rate (USD) column contained mixed values such as "USD 100", "$40", and numeric values.

*   Non-numeric characters were removed, and the column was converted into numeric format. Missing values were filled using the median due to possible outliers.



```python
df["hourly_rate (USD)"].head()
```

**Client Satisfaction**

```python
df["client_satisfaction"] = df["client_satisfaction"].astype(str)
df["client_satisfaction"] = df["client_satisfaction"].str.replace("%","")
df["client_satisfaction"] = pd.to_numeric(df["client_satisfaction"], errors="coerce")
df["client_satisfaction"].fillna(df["client_satisfaction"].median(), inplace=True)
```

The client_satisfaction column contained percentage values. The percentage symbol and extra spaces were removed, and the column was converted into numeric format. Missing values were replaced using the median.

```python
df["client_satisfaction"].head()
```

**Gender**

```python
df["gender"] = df["gender"].astype(str).str.strip().str.lower()

df["gender"] = df["gender"].replace({
    "m": "male",
    "f": "female"
})

df["gender"] = df["gender"].str.title()
```

The gender column was standardized by removing spaces and converting values into consistent categories such as Male and Female.

```python
df["gender"].unique()
```

**Is Active**

```python
df["is_active"] = df["is_active"].astype(str)

df["is_active"] = df["is_active"].replace({
    "1": "Yes",
    "Y": "Yes",
    "yes": "Yes",
    "True": "Yes",

    "0": "No",
    "N": "No",
    "no": "No",
    "False": "No"
})

df["is_active"] = df["is_active"].replace("nan", np.nan)

df["is_active"].fillna(df["is_active"].mode()[0], inplace=True)
```

```python
df["is_active"].unique()
```

The is_active column contained inconsistent values such as "1", "Y", "True", "0", "N", and "False", along with missing values.

First, the column was converted into string format to handle mixed data types. Then, all values were standardized into two consistent categories:
- "Yes" for active freelancers  
- "No" for inactive freelancers  

During conversion, missing values were represented as the string "nan". These were converted back into proper missing values using replace().

Finally, missing values were filled using the mode (most frequent value) to maintain data consistency and avoid introducing new categories.

This ensures that the column is clean, consistent, and suitable for analysis.

**Creating Derived Columns**

**Creating Experience Level Category**

```python
df["experience_level"] = pd.cut(
    df["years_of_experience"],
    bins=[-1, 2, 5, 10, 50],
    labels=["Beginner", "Intermediate", "Advanced", "Expert"]
)
```

A new column named experience_level was created to categorize freelancers based on their years of experience.

The pd.cut() function was used to divide the continuous numerical data into meaningful categories:
- Beginner (0–2 years)
- Intermediate (3–5 years)
- Advanced (6–10 years)
- Expert (10+ years)

This transformation simplifies analysis by grouping freelancers into levels, making it easier to compare performance and trends across experience categories.

```python
df["high_value_freelancer"] = df["hourly_rate (USD)"] > 50

df["high_value_freelancer"] = df["high_value_freelancer"].map({
    True: "Yes",
    False: "No"
})
```

A new column named high_value_freelancer was created to identify freelancers earning more than 50 USD per hour.

It is observed that freelancers with higher experience levels, particularly experts, tend to earn slightly higher hourly rates. However, the differences are not very large, indicating that factors such as skill type and market demand also play an important role in determining earnings.

The distribution of skills shows a balanced market, with DevOps, UI/UX Design, and Blockchain Development being slightly more prominent. Technical skills such as Cybersecurity, Web Development, and DevOps offer marginally higher earning potential, suggesting that specialized skills are valued in the freelancing industry.

The dataset also highlights a balanced gender distribution with no significant difference in earnings between male and female freelancers, indicating fairness in pay. Additionally, active freelancers tend to earn more and maintain better ratings, emphasizing the importance of engagement and consistency.

Country-wise analysis shows that freelancer participation varies across regions, although performance and ratings remain relatively consistent globally. The correlation analysis reveals that age and experience are moderately related, while other variables such as hourly rate, rating, and client satisfaction are influenced by multiple external factors.

Overall, the freelancing market appears to be competitive, diverse, and performance-driven, with experience, skills, and activity level acting as the key factors influencing success.
